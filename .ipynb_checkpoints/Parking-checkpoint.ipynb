```python
import pandas as pd
```

```python
# df containging all zip
df = pd.read_csv("311_Service_Requests_2020.csv")
df.head()
```

```python
# df2 is a subset of df, containing rows that include zip 10001
df2 = df[df["Incident Zip"] == 10001.0]
df2.head()
```

```python
complaints = list(df2["Complaint Type"])
total_counts = 0
parking_counts = 0

for i in complaints:
    if i == "Illegal Parking":
        parking_counts += 1
    total_counts += 1 

print("total counts:", total_counts)
print("parking counts:", parking_counts)
```

```python
my_zip_proportion = (parking_counts / total_counts) * 100
print(my_zip_proportion)
```

```python
total_complaints = list(df["Complaint Type"])
total_zip_counts = 0
total_parking_counts = 0

for i in total_complaints:
    if i == "Illegal Parking":
        total_parking_counts += 1
    total_zip_counts += 1 

print("total zip counts:", total_zip_counts)
print("total parking counts:", total_parking_counts)
```

```python
total_zip_proportion = (total_parking_counts / total_zip_counts) * 100
print(total_zip_proportion)
```

```python
higher_parking_proportion = my_zip_proportion > total_zip_proportion
print(higher_parking_proportion)
```
